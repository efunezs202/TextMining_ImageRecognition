### Normalización de Texto

```python
def normalize_text(corpus):
    newCorpus = []
    for doc in corpus:
        newCorpus.append(re.sub(r'[^a-zA-Z0-9\s{1}áéíóúüñÑÁÉÍÓÚ]', '', doc).lower().strip().rstrip('\r\n'))
                        
    return newCorpus
```

### Tokenización

```python
def tokenization (corpus):
    newCorpus = []
    for doc in corpus:
        newCorpus.append(nlp(doc))
    return newCorpus
```

### Remove Stopwords

```python
def removeStopWords (corpus):
    newCorpus = []
    for doc in corpus:
        s = ''
        for token in doc:
            if(token.is_stop == False):
                
                s = s + token.text + ' '
        
        newCorpus.append(s.strip())
    return newCorpus
```

### Stemming y Lemmatization

```python
def stemming_lemmatization (corpus):
    docs = tokenization(corpus)
    newCorpus = []
    for doc in docs:
        s = ''
        for token in doc:
            s = s + token.lemma_ + ' '
        newCorpus.append(s.strip())
    return newCorpus
```

```python
corpus_stemLem1 = stemming_lemmatization(corpus_remstp1)
```

```python
corpus_stemLem2 = stemming_lemmatization(corpus_remstp2)
```

```python
corpus_stemLem3 = stemming_lemmatization(corpus_remstp3)
```
